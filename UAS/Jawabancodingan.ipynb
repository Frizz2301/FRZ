```python
from datetime import datetime

class Kerjasama:
    def __init__(self, jenis_proyek, mitra_kerjasama, tanggal_mulai, tanggal_selesai, lokasi_proyek, tujuan_kerjasama):
        self.jenis_proyek = jenis_proyek
        self.mitra_kerjasama = mitra_kerjasama
        self.tanggal_mulai = datetime.strptime(tanggal_mulai, "%Y-%m-%d")
        self.tanggal_selesai = datetime.strptime(tanggal_selesai, "%Y-%m-%d")
        self.lokasi_proyek = lokasi_proyek
        self.tujuan_kerjasama = tujuan_kerjasama

    def info_kerjasama(self):
        info = f"""
        Jenis Proyek     : {self.jenis_proyek}
        Mitra Kerjasama  : {self.mitra_kerjasama}
        Tanggal Mulai    : {self.tanggal_mulai.strftime("%Y-%m-%d")}
        Tanggal Selesai  : {self.tanggal_selesai.strftime("%Y-%m-%d")}
        Lokasi Proyek    : {self.lokasi_proyek}
        Tujuan Kerjasama : {self.tujuan_kerjasama}
        """
        return info

def tambah_kerjasama():
    jenis_proyek = input("Masukkan jenis proyek: ")
    mitra_kerjasama = input("Masukkan mitra kerjasama: ")
    tanggal_mulai = input("Masukkan tanggal mulai (YYYY-MM-DD): ")
    tanggal_selesai = input("Masukkan tanggal selesai (YYYY-MM-DD): ")
    lokasi_proyek = input("Masukkan lokasi proyek: ")
    tujuan_kerjasama = input("Masukkan tujuan kerjasama: ")

    kerjasama = Kerjasama(jenis_proyek, mitra_kerjasama, tanggal_mulai, tanggal_selesai, lokasi_proyek, tujuan_kerjasama)
    return kerjasama

def main():
    daftar_kerjasama = []

    while True:
        print("1. Tambah Program Kerjasama")
        print("2. Lihat Program Kerjasama")
        print("3. Keluar")
        pilihan = input("Pilih opsi: ")

        if pilihan == "1":
            kerjasama = tambah_kerjasama()
            daftar_kerjasama.append(kerjasama)
            print("Program kerjasama berhasil ditambahkan!")
        elif pilihan == "2":
            if not daftar_kerjasama:
                print("Tidak ada program kerjasama yang terdaftar.")
            else:
                for i, kerjasama in enumerate(daftar_kerjasama, 1):
                    print(f"Program Kerjasama {i}:")
                    print(kerjasama.info_kerjasama())
        elif pilihan == "3":
            break
        else:
            print("Pilihan tidak valid. Silakan coba lagi.")

if __name__ == "__main__":
    main()
```
